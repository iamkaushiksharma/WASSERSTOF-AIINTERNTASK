## looping through every folder in our directory for currupted images or anything else that are not in classes listed above

```python
for image_class in os.listdir(data_dir):   #looping through every folder inside my data directory.
    for image in os.listdir(os.path.join(data_dir,image_class)):   #Looping to print every single image into our sub directory.
        image_path = os.path.join(data_dir, image_class, image)
        try:
            img = cv2.imread(image_path)   #loop to check i can load the imagr to opencv
            tip = imger.what(image_path)   #to check that all the matches the path above in image_ext list.
            if tip not in image_exts:
                print('image not in ext list {}'.format(image_path))
                os.remove(image_path)
        except Exception as e:
            print('issuse with image {}'.format(image_path))
```
